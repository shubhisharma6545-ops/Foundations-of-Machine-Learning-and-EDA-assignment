#1. What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

   - **Artificial Intelligence (AI)** is the broadest field that focuses on creating machines capable of performing tasks that typically require human intelligence, such as reasoning, problem-solving, perception, and decision-making. It combines statistics, data analysis, programming, and machine learning techniques. Its scope includes data collection, data cleaning, data visualization, statistical modeling, and predictive analysis. Data science is applied in business analytics, healthcare analysis, financial forecasting, customer behavior analysis, and decision-making support systems.

#2. Explain overfitting and underfitting in ML. How can you detect and prevent them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.
   - **Overfitting** and **underfitting** are common problems in **machine learning models** related to how well a model learns from training data and generalizes to new data.

   1. **Overfitting** occurs when a model learns the training data too well, including noise and random fluctuations. As a result, the model performs very well on the training dataset but poorly on unseen test data. This usually happens when the model is too complex for the amount of data available. Overfitting is characterized by **low bias and high variance**, meaning the model captures too many details that do not generalize well. It can be detected when the **training accuracy is very high but validation or test accuracy is significantly lower**. To prevent overfitting, techniques such as **cross-validation**, **regularization (L1 and L2)**, **dropout**, **reducing model complexity**, **early stopping**, and **increasing training data** can be used.

   2. **Underfitting**, on the other hand, occurs when a model is too simple to capture the underlying patterns in the data. In this case, the model performs poorly on both training and test datasets because it fails to learn the important relationships in the data. Underfitting is associated with **high bias and low variance**, meaning the model makes overly simplistic assumptions. It can be detected when **both training and validation errors are high**. To prevent underfitting, one can **increase model complexity**, **add more relevant features**, **train the model for longer**, or **use more advanced algorithms**.

The **bias–variance tradeoff** explains the balance between these two problems. A good model should have a balance where bias and variance are minimized to achieve better generalization. **Cross-validation**, such as **k-fold cross-validation**, helps detect overfitting and underfitting by evaluating the model on multiple subsets of the data, ensuring more reliable performance estimates. **Regularization techniques** like **L1 (Lasso)** and **L2 (Ridge)** add penalties to model parameters to prevent overly complex models and reduce overfitting.

#3. How would you handle missing values in a dataset? Explain at least three methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.
   - Handling **missing values** is an important step in data preprocessing because incomplete data can affect the performance and accuracy of machine learning models. There are several methods to deal with missing values depending on the dataset and problem.

**1. Deletion (Removing Missing Data)**
One simple approach is to remove the rows or columns that contain missing values. This method works well when the number of missing values is small and does not significantly affect the dataset size. There are two common types: **listwise deletion** (removing entire rows with missing values) and **column deletion** (removing a feature with many missing values). For example, if a dataset has 10,000 records and only 50 rows have missing values in the “Age” column, those rows can be removed without greatly affecting the analysis. However, if many records contain missing values, deletion may lead to loss of important information.

**2. Mean/Median Imputation**
In this method, missing values are replaced with the **mean, median, or mode** of the column. Mean imputation is commonly used for numerical data, while median is preferred when the data contains outliers. Mode is used for categorical data. For example, if the “Salary” column has missing values and the average salary is ₹50,000, the missing entries can be filled with ₹50,000. If the data is skewed, the median salary may be used instead. This method is simple and quick but may reduce data variability.

**3. Predictive Modeling (Model-Based Imputation)**
A more advanced method is to predict the missing values using machine learning models. In this approach, other features in the dataset are used to estimate the missing values. Techniques such as **regression, decision trees, or k-nearest neighbors (KNN)** can be used. For example, if the “Income” value is missing for some individuals, a regression model can predict it based on features like education level, age, and occupation. This approach often provides more accurate estimates but requires additional computation and careful model selection.

In practice, the choice of method depends on the **amount of missing data, the type of variables, and the importance of preserving information in the dataset**. Often, data scientists experiment with different methods and evaluate their impact on model performance.

#4. What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models
   - An **imbalanced dataset** is a dataset in which the classes are not equally represented. One class (called the **majority class**) has significantly more samples than the other class (called the **minority class**). This often occurs in real-world problems such as fraud detection, disease diagnosis, or spam detection. For example, in a credit card fraud dataset, 98% of the transactions may be legitimate while only 2% are fraudulent. In such cases, a machine learning model may become biased toward the majority class and achieve high accuracy while still failing to correctly identify the minority class.

One technique to handle an imbalanced dataset is **Random Oversampling**. In this method, the minority class samples are randomly duplicated to increase their number until the dataset becomes more balanced. The idea is to give the model more exposure to the minority class during training. For example, if a dataset contains 900 normal transactions and 100 fraudulent transactions, oversampling can duplicate fraud samples to make them closer to 900. Practically, this can be implemented in Python using libraries like `imblearn` with `RandomOverSampler`, which automatically replicates minority class samples.

Another technique is **SMOTE (Synthetic Minority Oversampling Technique)**. Instead of simply duplicating existing samples, SMOTE generates **synthetic examples** of the minority class by interpolating between existing minority samples. This helps reduce overfitting that can occur with simple duplication. For example, if two minority class data points are close in feature space, SMOTE can generate a new synthetic data point between them. In practice, this can be implemented using `SMOTE` from the `imblearn.over_sampling` module in Python to create balanced datasets before training a machine learning model.

Another practical approach is using **class weights** in machine learning models. Instead of modifying the dataset, this method assigns higher penalties to mistakes made on the minority class during training. For example, in models such as logistic regression, decision trees, or support vector machines, the `class_weight` parameter can be set so that misclassifying the minority class is penalized more heavily. This helps the model pay more attention to the minority class without changing the dataset itself.

#5. Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.
   - **Feature scaling** is an important preprocessing step in machine learning because different features in a dataset may have different ranges and units. If features are not scaled, those with larger numerical values can dominate the learning process and negatively affect the performance of the model. Feature scaling ensures that all features contribute equally to the model. It is especially important for **distance-based algorithms** such as KNN and SVM because these algorithms calculate distances between data points. If one feature has a much larger scale than others, it will heavily influence the distance calculation. Feature scaling also improves the efficiency of **gradient descent optimization** used in many algorithms by helping the model converge faster and more smoothly during training.

**Min-Max Scaling (Normalization)** is a technique that rescales the feature values to a fixed range, usually between **0 and 1**. It uses the formula:
[
X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}
]
This method preserves the original distribution of the data but changes its scale. For example, if a feature like income ranges from 10,000 to 100,000, Min-Max scaling converts it into values between 0 and 1. It works well when the data does not contain many outliers and when a bounded range is required.

**Standardization (Z-score normalization)** transforms the data so that it has a **mean of 0 and a standard deviation of 1**. It uses the formula:
[
X_{scaled} = \frac{X - \mu}{\sigma}
]
where ( \mu ) is the mean and ( \sigma ) is the standard deviation. Unlike Min-Max scaling, standardization does not limit the data to a specific range and is generally more robust when the data contains outliers. It is commonly used in algorithms like logistic regression, SVM, and neural networks because it helps gradient descent converge more efficiently.

In summary, **Min-Max scaling** rescales data to a fixed range (0–1), while **standardization** transforms data to have zero mean and unit variance. Both methods improve the performance of machine learning algorithms by ensuring that features are on a comparable scale.

#6. Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.
   - **Label Encoding** and **One-Hot Encoding** are techniques used to convert **categorical variables into numerical form** so that machine learning algorithms can process them.

   1. **Label Encoding** assigns a **unique integer value** to each category in a feature. For example, a “Size” feature with values **Small, Medium, Large** might be encoded as **Small = 0, Medium = 1, Large = 2**. This method is simple and uses less memory because it represents each category with a single number. Label encoding is most appropriate when the categorical variable has an **ordinal relationship**, meaning the categories have a natural order (e.g., education level: High School, Bachelor’s, Master’s, PhD). However, for categories without order, label encoding may mislead the model because the algorithm might assume that higher numbers represent greater importance or magnitude.

   2. **One-Hot Encoding**, on the other hand, creates **separate binary columns for each category**. Each category is represented by a column where the value is **1 if the record belongs to that category and 0 otherwise**. For example, a “Color” feature with values **Red, Blue, Green** would become three columns: **Color_Red, Color_Blue, Color_Green**. This method does not impose any numerical relationship between categories, making it suitable for **nominal variables** where there is **no inherent order**, such as city names, product types, or colors. However, one-hot encoding can increase the dataset size significantly if there are many categories.

In practice, **Label Encoding is preferred for ordinal categorical variables** where order matters, while **One-Hot Encoding is preferred for nominal categorical variables** where categories are independent and have no ranking.

```python
#7.

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Remove rows with missing ratings
df = df.dropna(subset=['Rating'])

# Convert rating column to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Calculate average rating by category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Display top and bottom categories
print("Top 5 Categories with Highest Ratings:")
print(category_ratings.head())

print("\nBottom 5 Categories with Lowest Ratings:")
print(category_ratings.tail())

# Visualization
plt.figure(figsize=(10,6))
category_ratings.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
#8.

# Import libraries
import pandas as pd

# Load dataset
df = pd.read_csv("titanic.csv")

# -------------------------------
# (a) Survival rate by Pclass
# -------------------------------

pclass_survival = df.groupby("Pclass")["Survived"].mean()

print("Survival Rate by Passenger Class:")
print(pclass_survival)

# -------------------------------
# (b) Age group analysis
# -------------------------------

# Remove missing ages
df_age = df.dropna(subset=["Age"])

# Create age groups
df_age["Age_Group"] = df_age["Age"].apply(lambda x: "Child" if x < 18 else "Adult")

age_survival = df_age.groupby("Age_Group")["Survived"].mean()

print("\nSurvival Rate by Age Group:")
print(age_survival)
```

```python
#9.

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("flight_price.csv")

# ---------------------------------
# (a) Price vs Days Left
# ---------------------------------

# Average price based on days left
price_by_days = df.groupby("days_left")["price"].mean()

print("Average Price by Days Left:")
print(price_by_days.head())

# Plot relationship
plt.figure(figsize=(8,5))
plt.plot(price_by_days.index, price_by_days.values)
plt.xlabel("Days Left Before Departure")
plt.ylabel("Average Flight Price")
plt.title("Flight Price vs Days Left Before Departure")
plt.show()


# ---------------------------------
# (b) Airline price comparison
# ---------------------------------

# Filter Delhi to Mumbai route
route_df = df[(df["source_city"]=="Delhi") & (df["destination_city"]=="Mumbai")]

# Average price per airline
airline_price = route_df.groupby("airline")["price"].mean().sort_values()

print("\nAverage Price by Airline (Delhi-Mumbai):")
print(airline_price)

# Bar chart
plt.figure(figsize=(8,5))
airline_price.plot(kind="bar", color="orange")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.title("Airline Price Comparison (Delhi-Mumbai)")
plt.show()
```

```python
#10.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("hr_analytics.csv")

# --------------------------
# (a) Correlation with Attrition
# --------------------------
# 'left' column already contains 0 for stayed and 1 for left, no mapping needed

# Compute correlations, considering only numeric columns
corr = df.corr(numeric_only=True)['left'].sort_values(ascending=False)
print("Correlation with Attrition:\n", corr)

# Visualize key drivers
plt.figure(figsize=(8,5))
sns.barplot(x=corr.index[1:6], y=corr.values[1:6])
plt.title("Top 5 Factors Correlated with Attrition")
plt.ylabel("Correlation with Attrition")
plt.xticks(rotation=45)
plt.show()

# --------------------------
# (b) Effect of Number of Projects
# --------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x='left', y='number_project', data=df)
plt.title("Number of Projects vs Attrition")
plt.xlabel("Attrition (0=Stayed, 1=Left)")
plt.ylabel("Number of Projects")
plt.show()

# Average projects for employees who left vs stayed
avg_projects = df.groupby('left')['number_project'].mean()
print("\nAverage Number of Projects:")
print(avg_projects)
```
